## Data Analysis

### Top 5 most popular artists

```python
top__five_artists = df['artist'].value_counts().head()
top__five_artists
```

```python
top__five_artists.plot(kind = 'bar',rot = 0,figsize = (8,6))
```

### Top Five Loudest Tracks 


```python
top_five_loudest_track = df[['loudness','song_title']].sort_values(by = 'loudness',ascending = True).head()
top_five_loudest_track
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='loudness',y='song_title',data = top_five_loudest_track)
plt.title('Top 5 loudest track')
plt.show()
```

### Artist with the most danceability song

```python
top_five_artist_danceability_songs = df[['danceability','song_title','artist']].sort_values(by='danceability',ascending = False).head()
top_five_artist_danceability_songs 
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='danceability',y='song_title',data = top_five_artist_danceability_songs )
plt.title('Artist with most danceability song')
plt.show()
```

### Top 10 Instrumentalness tracks

```python
top_ten_instrumental_tracks = df[['instrumentalness','song_title','artist']].sort_values(by='instrumentalness',ascending = False ).head()
top_ten_instrumental_tracks 
```

```python
plt.figure(figsize=(12, 7))


plt.pie(x=top_ten_instrumental_tracks['instrumentalness'], 
        autopct='%1.2f%%', 
        labels=top_ten_instrumental_tracks['song_title'])

plt.title('Top 10 Instrumental Tracks')
plt.show()
```

 ### Top 10 Energetic tracks

```python
top_ten_energetic_tracks = df[['energy','song_title','artist']].sort_values(by='energy',ascending = False ).head()
top_ten_energetic_tracks 
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='artist',y='energy',data = top_ten_energetic_tracks)
plt.title('Artist with most Energetic song')
plt.show()
```
